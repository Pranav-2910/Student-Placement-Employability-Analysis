**Performing data cleaning on the dataset**

```python
#Importing libraries
import pandas as pd
```

```python
#Importing dataset
df = pd.read_csv('/content/train.csv')
```

```python
#Checking null values
df.isnull().sum()
```

```python
#Encoding placement status
df['Placed'] = df['Placement_Status'].map({'Placed':1,'Not Placed':0})
```

```python
#Creating a column skill average
df['Avg_Skill_Score'] = (df['Coding_Skills'] + df['Communication_Skills'] + df['Soft_Skills_Rating']) / 3
```

```python
#Creating a column experience score
df['Experience_Score'] = (df['Internships']*2 + df['Projects'] + df['Certifications'])
```

```python
#Converting dataframe to csv
df.to_csv('cleaned_placement_data.csv', index=False)
```
